## Detailed view about how it works


```python
import pandas as pd
import numpy as np
```

### Loading the data set

```python
train_data = pd.read_csv("../input/random-linear-regression/train.csv")
test_data = pd.read_csv("../input/random-linear-regression/test.csv")
```

### Data Exploration

### Data cleaning

```python
train_data.isnull().sum()
```

```python
train_data_set=train_data.dropna()
```

```python
train_data_set.isnull().sum()
```

```python
test_data_set=test_data.dropna()
```

```python
test_data_set.isnull().sum()
```

### Data correlation

```python
train_data_set.corr()
```

```python
x_train=train_data_set['x']
y_train= train_data_set['y']
```

```python
x_test=test_data_set['x']
y_test=test_data_set['y']
```

### Handling outliers

**handling by standard deviation**

```python
data_std_X = x_train.std()
data_mean_X = x_train.mean()
```

```python
data_std_Y = y_train.std()
data_mean_Y = y_train.mean()
```

#### Caluculating outliers ...

```python
print(data_std_X,data_mean_X)
print(data_std_Y,data_mean_Y)

lower_x , upper_x = data_mean_X - 3* data_std_X , data_mean_X + 3* data_std_X
lower_y , upper_y = data_mean_Y - 3 * data_std_Y ,data_mean_Y + 3 * data_std_Y

print(lower_x , upper_x )
print(lower_y , upper_y)

outliers_std_x = [x for x in x_train if x <lower_x or x > upper_x]
outliers_std_y = [y for y in y_train if y <lower_y or y > upper_y]

print(outliers_std_x)
print(outliers_std_y)
```

## Linear Regression 

#### Reshaping the array

```python
train_x = x_train.values.reshape(-1,1)
train_y = y_train.values.reshape(-1,1)

test_x = x_test.values.reshape(-1,1)
test_y = y_test.values.reshape(-1,1)
```

### Using sklearn

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_x,train_y)
```

Predict the test data

```python
y_predict= lr.predict(test_x)
test_data_set['predicted'] = y_predict
```

```python
test_data_set
```

### Model Evaluation for performance

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(test_x,test_y,color="red")
plt.plot(test_x,y_predict)
```

### caluculating mean_squared_error

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(mean_squared_error(test_y,y_predict))
print(mean_absolute_error(test_y,y_predict))
print(r2_score(test_y,y_predict))
```
